```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
AGraph = nx.Graph() # cria um grafo com networkx
Nodes = range(1, 5) # cria os nós (nodes), que vai de 1 à 5
Edges = [(1, 2), (2, 3), (3, 4), (4, 5), (1, 3), (1, 5)] # cria as arestas (edges), que conectam os nós
AGraph.add_nodes_from(Nodes) # adicionam os nós (Nodes) ao grafo AGraph
AGraph.add_edges_from(Edges) # adicionam as arestas (Edges) ao grafo AGraph
```

```python
nx.draw(AGraph, with_labels=True)
```
